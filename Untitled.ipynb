```python
# This calculates the difference between timers when writen as string
def calculate_time(time1, time2):
    result_sec = 0
    result_min = 0
    result_hour = 0
    
    # Converting from string to int
    try: 
        hour1 = int(time1[:2])
        hour2 = int(time2[:2])
        min1 = int(time1[3:5])
        min2 = int(time2[3:5])
        sec1 = int(time1[6:8])
        sec2 = int(time2[6:8])
    except:
        return 0   #There are some celles that only has ":" as time so I decide to just give this a 0 as there is no good answer.

    result_sec = sec2 - sec1
    result_min = min2 - min1
    result_hour = hour2 - hour1
    
    if result_sec < 0:
        result_sec +=60
        result_min -=1
    if result_min < 0:
        result_min +=60
        result_hour -=1
    if result_hour < 0:
        result_hour +=24
        
    # Converts to seconds
    final_res = result_sec + (result_min * 60) + (result_hour * 3600)
    return final_res

def convert_to_seconds(time):
    try: 
        hour = int(time[:2])
        mins = int(time[3:5])
        sec = int(time[6:8])
    except:
        return 100
    return (hour*3600)+(mins*60)+sec
```

```python
# Here I am gonna calculate the difference between Tidspunkt_Faktisk_Ankomst_Holdeplass_Fra and 
#  Tidspunkt_Faktisk_Avgang_Holdeplass_Fra, and place how many seconds in a new array.

time_used_on_stop = []
i=0
while i < len(ruter_data_array):
    time_used_on_stop.append(calculate_time(ruter_data_array[i][11], ruter_data_array[i][12]))
    i+=1
```

```python
#I couldn't find a way to easily just use my time_used_on_stop array to make a visualization so I am making
# a new array that I am making into another DataFrame object so that I can use that to make a visualization instead.
time_of_day = []
i=0
while i < len(ruter_data_array):
    time_of_day += [convert_to_seconds(ruter_data_array[i][13])]
    i+=1
# I converted everything into seconds because that was the easiest way I found to give an easy "scalable" way to show time
#  as it didn't work as a string. It still works fine as a visualization as you still see if it is early or late. 
finding_when_stop_time_happens = [[]]

y=0
while y < len(time_used_on_stop):
    finding_when_stop_time_happens.append([time_of_day[y],time_used_on_stop[y], calculate_time(ruter_data_array[y][12], ruter_data_array[y][14])])
    y+=1
```

```python
finding_standstill_times = pd.DataFrame(finding_when_stop_time_happens, columns=["Time_Of_Day", "Spent_On_Stop", "Variance"])
finding_standstill_times.plot.scatter(x="Time_Of_Day", y="Spent_On_Stop", title="Standing still over a day")
```

```python
# Making a list containing the TurID, the maximum capacity and the on board count.
highest_number_of_passengers = []
i=0
while i < len(ruter_data_array):
    highest_number_of_passengers += [[ruter_data_array[i][9], ruter_data_array[i][15], ruter_data_array[i][16]]]
    i+=1
```

```python
# Here I am trying to find the highest amount of passengers that are found on any route. Maybe I should have differed
# between "Linjeretningen", but I ended up not doing that as it probably doesn't matter too much, but could be something
# to think about. 
counted = []
i=0
while i < len(highest_number_of_passengers):
    if highest_number_of_passengers[i][0] not in counted:
        y=0
        current_high = highest_number_of_passengers[i][2]
        while y < len(highest_number_of_passengers):
            if (highest_number_of_passengers[y][0] == highest_number_of_passengers[i][0]) and (highest_number_of_passengers[y][2] > current_high):
                current_high = highest_number_of_passengers[y][2]    
            y+=1
        counted += [[highest_number_of_passengers[i][0], highest_number_of_passengers[i][1], current_high]]
        
    i+=1
# Now there should be made an array with a unique TurID, and it still contains its max capacity and now has a highscore.
```
